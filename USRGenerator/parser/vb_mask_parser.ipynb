```python
#this code is for crreating a mask and unmask sentence dataset.
import os
with open("10k_after_clean.txt","r") as f:
    unmasked_dataset=[]
    for sent in f:
        unmasked_dataset.append(sent)
for sentence in unmasked_dataset:
    try:
        with open("txt_files/bh-1","w") as f:
            f.write(sentence)
        os.system("isc-parser -i txt_files/bh-1 >hellow.hellow.txt ")
        t=open("txt_files/bh-1","r")
        orig_sent=t.readline()
        orig_sent=orig_sent.split()
        orig_sent_disp=orig_sent.copy()
        with open("hellow.hellow.txt","r") as f:
            parser_op=[]
            for line in f:
                parser_op.append(line.strip().split())
        pos_tag={}
        for op in range(len(parser_op)-1):
            pos_tag[parser_op[op][1]]=parser_op[op][3]
            #print(parser_op[op][3])
        for word in range(len(orig_sent)):
            #print(orig_sent[word])
            if pos_tag[orig_sent[word]]=="PSP":
                orig_sent_disp[word]="<mask>"

            elif pos_tag[orig_sent[word]]=="NST":
                psp_temp_list=["का","के","की"]
                if pos_tag[orig_sent[word-1]]=="PSP" and orig_sent[word-1] in psp_temp_list:
                    orig_sent_disp[word]="<mask>"
        sent1=" ".join([str(word) for word in orig_sent_disp])
        #f=open("masked_vib_dataset.txt","w")
        #f.write(sent1)
        t_1=open("txt_files/bh-1","r")
        orig_sent_print=t_1.readline()
        print(sent1+"\t"+orig_sent_print)
    except:
        continue
```
